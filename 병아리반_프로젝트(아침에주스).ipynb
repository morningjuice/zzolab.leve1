```python
import csv
import time
import matplotlib.pyplot as plt

data = csv.reader(open('greengas1990-2018.csv', encoding = 'cp949'))

#빈 리스트 만들기
x=[]                        #자료를 한 줄로 불러 오기위한 리스트
total = []                  #순배출량 리스트   
total1 = []                 #순배출량을 푼 값을 저장할 리스트 
list_x = []                 #년도를 저장할 리스트
period = []                 #년도를 푼 값을 저장할 리스트 

print('기후위기의 주범인 온실가스...')
time.sleep(1)
print('1990년도부터 2018년까지 얼마나 우리가 배출하고 있을까요?')
time.sleep(1)
print('--------------------------------------------------------------')
time.sleep(1)
print('원하는 년도의 입력하면 온실가스 배출량을 확인할 수 있습니다.')
year = input('[입력] 년도를 네 자리로 입력해주세요. :   ')
time.sleep(0.5)
print('--------------------------------------------------------------')

#필요한 수치를 자료로 한 줄 씩 불러오기
for row in data:
  x.append(row[3:])
  #print(x[-1])

#자료를 불러온 x에서 가장 마지막 값인 순배출량만 total 리스트에 저장
total.append(x[-1])

#total 리스트에 들어간 값을 풀어서 total1에 저장
for i in range(29):
  total1.append(float(total[0][i]))

#자료를 불러온 x에서 가장 첫번째 값인 년도만 list_x 리스트에 저장
list_x.append(x[0])

#list_x에 들어간 값을 풀어서 period에 저장
for i in range(29):
  period.append(list_x[0][i])

energy = []
energy1 = []
energy.append(x[7])
for i in range(29):
  energy1.append(float(energy[0][i]))

indus = []
indus1 = []
indus.append(x[8])
for i in range(29):
  indus1.append(float(indus[0][i]))

agri = []
agri1 = []
agri.append(x[9])
for i in range(29):
  agri1.append(float(agri[0][i]))

waste = []
waste1 = []
waste.append(x[10])
for i in range(29):
  waste1.append(float(waste[0][i]))

for i in range(29):
  if year == period[i]:
    print(year +'년도의 온실가스 순배출량은 ' + str(total1[i]) + '백만톤 CO2eq입니다.')

time.sleep(1)
print('--------------------------------------------------------------')
time.sleep(1)
print('1990년부터 2018년까지의 온실가스 순 배출량 및 구성을 그래프로 확인하고 싶으신가요?')
time.sleep(1)
answer = input('[입력] 그래프 확인을 원하면 1을 아니면 2를 입력하세요:   ')

if answer == '1':
  plt.figure(figsize = (25, 15))
  plt.plot(period, total1, marker='o', markerfacecolor='blue', markersize=6)
  plt.title('온실가스 순배출량(1990년~2018년)', fontsize = 20)
  plt.xlabel('연도', fontsize = 20)
  plt.ylabel('배출량', fontsize = 20)
  plt.show()

  plt.figure(figsize = (25, 15))
  plt.title('온실가스 구성', fontsize = 20)
  plt.plot(period, energy1, '-*', color='tomato', label = '에너지')
  plt.plot(period, indus1, '-o', color='gray', label = '산업공정')
  plt.plot(period, agri1, '-.', color='seagreen', label = '농업')
  plt.plot(period, waste1, marker = '^', markersize = 3, color='sienna', label = '폐기물')
  plt.legend()
  plt.show()
  # plt.figure(figsize = (20, 15))
  # ax1 = plt.subplot(4, 1, 1)
  # plt.plot(period, energy1, '.-', color='tomato')
  # plt.title('에너지')
  # plt.xticks(visible=False)

  # plt.figure(figsize = (20, 15))
  # plt.subplot(4, 1, 2, sharex=ax1)
  # plt.plot(period, indus1, '.-', color='gray')
  # plt.title('산업공정')
  # plt.xticks(visible=False)

  # plt.figure(figsize = (20, 15))
  # plt.subplot(4, 1, 3, sharex=ax1)
  # plt.plot(period, agri1, '.-', color='seagreen')
  # plt.title('농업')
  # plt.xticks(visible=False)

  # plt.figure(figsize = (20, 15))
  # plt.subplot(4, 1, 4, sharex=ax1)
  # plt.plot(period, waste1, '.-', color='sienna')
  # plt.title('폐기물')

  # plt.show()
```
